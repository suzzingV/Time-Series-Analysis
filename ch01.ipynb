```r
## Figure 1.1 : 불규칙성분

set.seed(1245)
n = 100
z <- 5000 + 20 * rnorm(n) # 정규 난수 생성
z.ts <- ts(z, start = c(1980, 1), frequency = 12) # 시계열 자료로 변환
ts.plot(z.ts, xlab = "date", ylab = "Zt", main = "그림 1-1 불규칙성분"); abline(h = 5000)
```

```r
## Figure 1.2 : 추세성분

set.seed(1234)
n = 100; t <- 1:n
x <- 0.5 * t
z <- 0.5 * t + rnorm(n)
z.ts <- ts(z, start = c(1980, 1), frequency = 12)
x.ts <- ts(x, start = c(1980, 1), frequency = 12)
ts.plot(z.ts, x.ts, col = c("blue", "red"), lty = 1:2, xlab = "date", ylab = "Zt", main = "그림 1-2 추세성분")
legend("topleft", legend = c("series", "trend"), lty = 1:2)
```

```r
## Figure 1.3 : 계절성분

n = 120; t <- 1:n
a <- rnorm(n, 0, 1)
z <- 10 + 3 * sin((2 * pi * t)/12) + 0.8 * a
z.ts <- ts(z, start = c(1985, 1), frequency = 12)
plot(z.ts, xlab = "date", ylab = "Zt", main = "그림 1-3 계절성분")
```

```r
## Figure 1.4 : 추세성분과 계절성분

z <- scan("/content/depart.txt")
lz <- log(z) # 추세성분 + 계절성분 -> 로그 변환
t <- 1:length(z)
x <- 6.3 + 0.012 * t
fig <- data.frame(lz, x) # 여러 개의 그림을 동시에 그리기 위해 data frame 생성
z.ts <- ts(fig, start = c(1984, 1), frequency = 12)
ts.plot(z.ts, lty = 1:2, xlab = "date", ylab = "logZ", main = "그림 1-4 추세성분과 계절성분")
legend("topleft", legend = c("series", "trend"), lty = 1:2)
```

```r
## Figure 1.5 : 변동폭이 다른 시계열자료
z <- scan("/content/koreapass.txt")
z.ts <- ts(z, start = c(1981, 1), frequency = 12)
ts.plot(z.ts, xlab = "date", ylab = "Zt", main = "그림 1-5 분산이 변하는 시계열")
```

```r
## Figure 1.6 : 추세선이 두 개인 시계열자료

set.seed(4321)
n = 120; t <- 1:n
a <- rnorm(n)
x <- 3 * (t - 46)
x[t<=60] = 0.5 * t[t<=60]
z <- x + a
z.ts <- ts(z, frequency = 12, start = c(1985, 1))
ts.plot(z.ts, xlab = "date", ylab = "Zt", main = "그림 1-6 추세가 변하는 시계열"); abline(v=1990)
```
